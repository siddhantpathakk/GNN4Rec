```python
import matplotlib.pyplot as plt
import re
import numpy as np
```

```python
data = {}
pattern = r'\d+\.\d+'
betas = [0.9563751300275105, 0.9564573957941945, 0.9370076133793354, 0.9168955733415564, 0.9599502484758489]
```

```python
import re
pattern = r'\d+\.\d+'

with open('betas.txt', 'r') as file:
    log_text = file.readlines()

loss, f1, precision, recall, ndcg = [], [], [], [], []

for line in log_text:
    if "end" in line:
        break
    elif "Loss" in line:
        float_values = [float(value) for value in re.findall(pattern, line)]
        loss.append(float_values[0])
        f1.append(float_values[1])
        precision.append(float_values[2])
        recall.append(float_values[3])
        ndcg.append(float_values[4])

assert len(loss) == len(f1) == len(precision) == len(recall) == len(ndcg) ==75
```

```python
data[str(betas[4])] = [loss, f1, precision, recall, ndcg]
```

```python
data.keys()
```

```python
# plot the loss of all the betas in one graph
plt.figure(figsize=(10, 8))
plt.title("Loss vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.plot(data[str(betas[0])][0], label=f"beta = {betas[0]}")
plt.plot(data[str(betas[1])][0], label=f"beta = {betas[1]}")
plt.plot(data[str(betas[2])][0], label=f"beta = {betas[2]}")
plt.plot(data[str(betas[3])][0], label=f"beta = {betas[3]}")
plt.plot(data[str(betas[4])][0], label=f"beta = {betas[4]}")
plt.legend()
plt.savefig("loss_vs_epochs.png")
```
