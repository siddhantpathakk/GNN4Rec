```python
import matplotlib.pyplot as plt
import re
```

```python
data = {}
```

```python
pattern = r'\d+\.\d+'
ds = ['amazonbooks', 'ml1m', 'yelp18']
for d in ds:
    with open(d+".txt", 'r') as file:
        log_text = file.readlines()
    epochs= []
    loss, f1, precision, recall, ndcg = [], [], [], [], []
    for line in log_text:
        if "epoch" in line:
            epochs.append(int(re.findall(r'\d+', line)[0]))
        if "Loss" in line:
            float_values = [float(value) for value in re.findall(pattern, line)]
            loss.append(float_values[0])
            f1.append(float_values[1])
            precision.append(float_values[2])
            recall.append(float_values[3])
            ndcg.append(float_values[4])
            
    data[d] = [loss, f1, precision, recall, ndcg, epochs]
```

```python
for d in ds:
    print("Number of epochs for", d, ":", len(data[d][5]))
```

```python
# For amazonbooks, plot loss in graph and the other metrics in subplot
plt.figure(figsize=(6, 6))
plt.plot(data['amazonbooks'][5], data['amazonbooks'][0])
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.savefig("amazonbooks_loss.png")

# plot the other metrics in subplot
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('amazonbooks')
axs[0, 0].plot(data['amazonbooks'][5], data['amazonbooks'][1])
axs[0, 0].set_title('F1')
axs[0, 1].plot(data['amazonbooks'][5], data['amazonbooks'][2])
axs[0, 1].set_title('Precision')
axs[1, 0].plot(data['amazonbooks'][5], data['amazonbooks'][3])
axs[1, 0].set_title('Recall')
axs[1, 1].plot(data['amazonbooks'][5], data['amazonbooks'][4])
axs[1, 1].set_title('NDCG')
plt.savefig('amazonbooks_metrics.png')
```
