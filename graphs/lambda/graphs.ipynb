```python
import matplotlib.pyplot as plt
import re
import numpy as np
```

```python
data = {}
pattern = r'\d+\.\d+'
lambdas = [0,1e-1,1e-2,1e-3,1e-4]
```

```python
import re
pattern = r'\d+\.\d+'

with open('lambda.txt', 'r') as file:
    log_text = file.readlines()

loss, f1, precision, recall, ndcg = [], [], [], [], []

for line in log_text:
    if "completed" in line:
        break
    elif "Loss" in line:
        float_values = [float(value) for value in re.findall(pattern, line)]
        loss.append(float_values[0])
        f1.append(float_values[1])
        precision.append(float_values[2])
        recall.append(float_values[3])
        ndcg.append(float_values[4])

assert len(loss) == len(f1) == len(precision) == len(recall) == len(ndcg) == 75
```

```python
data[str(lambdas[4])] = [loss, f1, precision, recall, ndcg]
```

```python
data.keys()
```

```python
# plot the loss of all the lambdas in one graph
plt.figure(figsize=(10, 8))
plt.title("Loss vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.plot(data[str(lambdas[0])][0], label="lambda = 0")
plt.plot(data[str(lambdas[1])][0], label="lambda = 1e-1")
plt.plot(data[str(lambdas[2])][0], label="lambda = 1e-2")
plt.plot(data[str(lambdas[3])][0], label="lambda = 1e-3")
plt.plot(data[str(lambdas[4])][0], label="lambda = 1e-4")
plt.legend()
```

```python
# plot the loss of all the lambdas in one graph
plt.figure(figsize=(10, 8))
plt.title("F1 vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("F1")
plt.plot(data[str(lambdas[0])][1], label="lambda = 0")
plt.plot(data[str(lambdas[1])][1], label="lambda = 1e-1")
plt.plot(data[str(lambdas[2])][1], label="lambda = 1e-2")
plt.plot(data[str(lambdas[3])][1], label="lambda = 1e-3")
plt.plot(data[str(lambdas[4])][1], label="lambda = 1e-4")
plt.legend()
```
