```python
import matplotlib.pyplot as plt
import re
import numpy as np
```

```python
data = {}
pattern = r'\d+\.\d+'
gammas = [0,1e-1,1e-2,1e-3,1e-4]
```

```python
import re
pattern = r'\d+\.\d+'

with open('gammas.txt', 'r') as file:
    log_text = file.readlines()

loss, f1, precision, recall, ndcg = [], [], [], [], []

for line in log_text:
    if "end" in line:
        break
    elif "Loss" in line:
        float_values = [float(value) for value in re.findall(pattern, line)]
        loss.append(float_values[0])
        f1.append(float_values[1])
        precision.append(float_values[2])
        recall.append(float_values[3])
        ndcg.append(float_values[4])

assert len(loss) == len(f1) == len(precision) == len(recall) == len(ndcg) ==75
```

```python
data[str(gammas[4])] = [loss, f1, precision, recall, ndcg]
```

```python
data.keys()
```

```python
# plot the loss of all the gammas in one graph
plt.figure(figsize=(10, 8))
plt.title("Loss vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.plot(data[str(gammas[0])][0], label=f"gamma = {gammas[0]}")
plt.plot(data[str(gammas[1])][0], label=f"gamma = {gammas[1]}")
plt.plot(data[str(gammas[2])][0], label=f"gamma = {gammas[2]}")
plt.plot(data[str(gammas[3])][0], label=f"gamma = {gammas[3]}")
plt.plot(data[str(gammas[4])][0], label=f"gamma = {gammas[4]}")
plt.legend()
plt.savefig("loss_vs_epochs.png")
```
